```python
import turtle
import math

tl = turtle.Turtle()

tl.penup()
tl.goto(-200,100)
tl.fillcolor('blue')

tl.pendown()
tl.right(90)
tl.forward(200)
tl.right(180)

tl.begin_fill()
for i in range (3):
    tl.pendown()
    tl.right(90)
    tl.forward(400)
    tl.left(90)
    tl.forward(40)
    tl.left(90)
    tl.forward(400)
    tl.right(90)
    tl.penup()
    tl.forward(40)
tl.end_fill()

tl.right(180)
tl.forward(40)
tl.fillcolor('red')
tl.begin_fill()
tl.pendown()
tl.left(60)
for i in range (3):
    tl.forward(200)
    tl.right(120)
tl.end_fill()

tl.penup()
tl.left(120)
tl.backward(120)
tl.right(90)
tl.forward(45)
tl.left(36)
tl.pendown()
tl.pencolor('white')
tl.fillcolor('white')
tl.begin_fill()
for i in range (5):
    tl.forward(55)
    tl.left(144)
tl.end_fill()

tl.penup()
tl.forward(300)

turtle.done()
```
