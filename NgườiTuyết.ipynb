```python
import turtle

tl = turtle.Turtle()

tl.penup()
tl.goto(0,-200)
tl.pendown()
#hình tròn thân
tl.fillcolor('#f0f8ff')
tl.begin_fill()
tl.circle (150,360)
tl.circle (150,180)
tl.end_fill()
#hình tròn đầu
tl.fillcolor('#faebd7')
tl.begin_fill()
tl.circle (-80,360)
tl.end_fill()
#Cúc áo 1
tl.left(90)
tl.penup()
tl.forward(220)
tl.pendown()
tl.right(90)
tl.fillcolor('black')
tl.begin_fill()
tl.circle (10,360)
tl.end_fill()
#Cúc áo 2
tl.right(90)
tl.penup()
tl.forward(60)
tl.left(90)
tl.fillcolor('black')
tl.begin_fill()
tl.circle (10,360)
tl.end_fill()
#Cúc áo 3
tl.right(90)
tl.penup()
tl.forward(60)
tl.left(90)
tl.fillcolor('black')
tl.begin_fill()
tl.circle (10,360)
tl.end_fill()
#Mũi
tl.right(90)
tl.penup()
tl.forward(190)
tl.left(90)
tl.fillcolor('red')
tl.begin_fill()
tl.circle (10,360)
tl.end_fill()
#Mắt trái
tl.penup()
tl.forward(30)
tl.right(90)
tl.penup()
tl.forward(15)
tl.left(90)
tl.fillcolor('black')
tl.begin_fill()
tl.circle (7,360)
tl.end_fill()
#mắt phải
tl.backward(60)
tl.fillcolor('black')
tl.begin_fill()
tl.circle (7,360)
tl.end_fill()
#miệng
tl.penup()
tl.forward(30)
tl.left(90)
tl.forward(70)
tl.left(90)
tl.pensize(10)
tl.pendown()
tl.circle (60,45)
tl.penup()
tl.circle (60,270)
tl.pendown()
tl.circle (60,45)
#mũ
tl.pensize(1)
tl.right(90)
tl.penup()
tl.backward(90)
tl.left(90)
tl.pendown()
tl.fillcolor('black')
tl.begin_fill()
tl.forward(70)
tl.left(120)
tl.forward(30)
tl.right(30)
tl.forward(10)
tl.circle (60,180)
tl.forward(10)
tl.right(30)
tl.forward(30)
tl.left(120)
tl.forward(90)
tl.end_fill()
#dây mũ
tl.penup()
tl.left(90)
tl.forward(30)
tl.left(90)
tl.pendown()
tl.pensize(15)
tl.pencolor('#ff7f50')
tl.forward(70)
tl.backward(110)
#Khăn
tl.penup()
tl.left(90)
tl.forward(155)
tl.right(90)
tl.pensize(40)
tl.pencolor('orange')
tl.pendown()
tl.forward(140)
tl.backward(20)
tl.left(70)
tl.pensize(35)
tl.forward(140)
tl.penup()
tl.backward(140)
tl.pendown()
tl.pensize(35)
tl.left(20)
tl.forward(160)
#Tay trái
tl.penup()
tl.right(150)
tl.forward(75)
tl.pensize(10)
tl.pendown()
tl.forward(100)
tl.right(30)
tl.pensize(6)
tl.forward(30)
tl.backward(30)
tl.pensize(10)
tl.left(30)
tl.forward(15)
tl.pensize(6)
tl.left(45)
tl.forward(30)
tl.backward(30)
tl.right(45)
tl.forward(30)
#Tay phải
tl.penup()
tl.backward(150)
tl.right(120)
tl.forward(220)
tl.left(40)
tl.pendown()
tl.pensize(10) 
tl.forward(120)
tl.left(40)
tl.pensize(6) 
tl.forward(30)
tl.backward(30)
tl.right(40)
tl.pensize(10) 
tl.forward(10)
tl.right(40)
tl.pensize(6) 
tl.forward(30)
tl.backward(30)
tl.left(40)
tl.pensize(10) 
tl.forward(35)

turtle.done()
```
